```python
import pandas as pd

# Step 1: Load CSV (treat empty strings and spaces as NaN)
df = pd.read_csv('/content/final_Rider.csv', na_values=["", " "])

# Step 2: Print original shape and column names
print("Original shape:", df.shape)
print("Original columns:", df.columns.tolist())

# Step 3: Clean column names
df.columns = df.columns.str.strip()

# Step 4: Drop rows where all columns are NaN
df = df.dropna(how='all')

# Step 6: Drop duplicate rows
df = df.drop_duplicates()

# Step 7: Print cleaned shape and preview
print("Cleaned shape:", df.shape)
df.head()
```

```python

output_excel_path = "/content/final_Rider.xlsx"
df.to_excel(output_excel_path, index=False)
print(f"Excel saved to: {output_excel_path}")
```
